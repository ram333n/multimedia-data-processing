# Task 2: Canny edge detection

```python
import os

import cv2
import matplotlib.pyplot as plt
```

```python
IMAGES_FOLDER = os.path.join(".", "images", "task_2")
```

```python
def show_image(image, title, cmap):
    plt.figure(figsize=(5, 5))
    plt.imshow(image, cmap=cmap)
    plt.title(title)
    plt.axis('off')
    plt.show()
```

```python
image = cv2.imread(os.path.join(IMAGES_FOLDER, "test_image.png"))
image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
show_image(image, title="Test image", cmap="gray")
```

```python
def show_canny_results(image, title, thresholds):
    n_thresholds = len(thresholds)
    plt.figure(figsize=(4 * n_thresholds, 4))
    plt.suptitle(title, fontsize=16, y=1.05)
    
    for i, (low_threshold, high_threshold) in enumerate(thresholds):
        edges = cv2.Canny(image, low_threshold, high_threshold)
        plt.subplot(1, n_thresholds, i + 1)
        plt.imshow(edges, cmap='gray')
        plt.title(f'Canny: low={low_threshold}, high={high_threshold}')
        plt.axis('off')
    
    plt.tight_layout()
    plt.show()
```

```python
show_canny_results(
    image,
    title="Canny with fixed upper threshold=200",
    thresholds=[(50, 200), (100, 200), (150, 200), (200, 200)]
)
```

```python
blurred_image = cv2.GaussianBlur(image, ksize=(5, 5), sigmaX=0) # sigma=0 -> OpenCV computes internally 
show_canny_results(
    blurred_image,
    title="Canny with Gaussian blur(5x5, sigma=0) and fixed upper threshold=200",
    thresholds=[(50, 200), (100, 200), (150, 200), (200, 200)]
)
```

```python
show_canny_results(
    image,
    title="Canny with fixed lower threshold=50",
    thresholds=[(50, 50), (50, 100), (50, 150), (50, 200)]
)
```

```python
show_canny_results(
    blurred_image,
    title="Canny with Gaussian blur(5x5, sigma=0) and fixed lower threshold=50",
    thresholds=[(50, 50), (50, 100), (50, 150), (50, 200)]
)
```
