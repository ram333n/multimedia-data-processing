# Task 3: Morphological operations (Erosion, Dilation, Opening, Closing)

```python
import os

import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
IMAGES_FOLDER = os.path.join(".", "images", "task_3")
```

```python
def show_image(image, title, cmap):
    plt.figure(figsize=(5, 5))
    plt.imshow(image, cmap=cmap)
    plt.title(title)
    plt.axis('off')
    plt.show()
```

```python
image = cv2.imread(os.path.join(IMAGES_FOLDER, "test_image.png"))
image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
show_image(image, title="Test image", cmap="gray")
```

```python
_, binary_image = cv2.threshold(image, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

show_image(binary_image, title="Binary image(Otsu)", cmap="gray")
```

```python
kernel = np.ones((3, 3), np.uint8)
```

```python
eroded_image = cv2.erode(binary_image, kernel=kernel, iterations=2)
show_image(eroded_image, title="Eroded image(kernel=3x3), iterations=2", cmap="gray")
```

```python
dilated_image = cv2.dilate(binary_image, kernel=kernel, iterations=2)
show_image(dilated_image, title="Dilated image(kernel=3x3), iterations=2", cmap="gray")
```

```python
opening_image = cv2.morphologyEx(binary_image, cv2.MORPH_OPEN, kernel, iterations=3)
show_image(opening_image, title="Opening result image(kernel=3x3), iterations=3", cmap="gray")
```

```python
closing_image = cv2.morphologyEx(binary_image, cv2.MORPH_CLOSE, kernel, iterations=3)
show_image(closing_image, title="Closing result image(kernel=3x3), iterations=3", cmap="gray")
```
